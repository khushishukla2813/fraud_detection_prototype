```python
data.drop(columns = ['age', 'total_claim_amount'], inplace = True, axis = 1)
```

```python
data.head()
```

```python
# Get Target and Indpendent Features Seperated
X = data.drop('fraud_reported', axis = 1)
y = data['fraud_reported']
```

```python
# Converting Label Columns into Numerical by doing One-Encoding
categorical_cols = X.select_dtypes(include = ['object'])
categorical_cols = pd.get_dummies(categorical_cols, drop_first = True)
categorical_cols.head()
```

```python
numerical_col = X.select_dtypes(include = ['int64'])
X = pd.concat([numerical_col, categorical_cols], axis = 1)
```

```python
X.head()
```

```python
# Outlier Check

plt.figure(figsize = (20, 15))
plotnumber = 1

for col in X.columns:
    if plotnumber <= 24:
        ax = plt.subplot(5, 5, plotnumber)
        sns.boxplot(X[col])
        plt.xlabel(col, fontsize = 15)
    
    plotnumber += 1
plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

```python
X_train.head()
```

```python
numerical_data = X_train[['months_as_customer', 'policy_deductable', 'umbrella_limit',
       'capital-gains', 'capital-loss', 'incident_hour_of_the_day',
       'number_of_vehicles_involved', 'bodily_injuries', 'witnesses', 'injury_claim', 'property_claim',
       'vehicle_claim']]
```

```python
# Standardization

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(numerical_data)
```

```python
scaled_num_df = pd.DataFrame(data = scaled_data, columns = numerical_data.columns, index = X_train.index)
scaled_num_df.head()
```

```python
import pandas as pd
import numpy as np
import xgboost as xgb
import joblib  

# Load trained model
model_path = "xgboost_model.pkl"  # Change to your actual model path
loaded_xgb_model = joblib.load(model_path)

# Load dataset
file_path = "fedu.csv"  # Change to your actual dataset path
df = pd.read_csv(file_path)

# Drop non-numeric columns (Modify as needed)
df = df.select_dtypes(include=[np.number])

# Convert DataFrame to NumPy array
fraud_samples = df.to_numpy()

# Ensure input shape matches model
expected_features = loaded_xgb_model.n_features_in_
actual_features = fraud_samples.shape[1]

if actual_features != expected_features:
    print(f"Feature mismatch! Expected {expected_features}, but got {actual_features}. Adjusting input...")
    if actual_features < expected_features:
        fraud_samples = np.pad(fraud_samples, ((0, 0), (0, expected_features - actual_features)), mode='constant')
    else:
        fraud_samples = fraud_samples[:, :expected_features]

# Make predictions
predictions = loaded_xgb_model.predict(fraud_samples)

# Convert numeric predictions to readable labels
predicted_labels = ["Fraud Detected" if pred == 1 else "Legitimate Transaction" for pred in predictions]

# Print results
for i, label in enumerate(predicted_labels):
    print(f"Sample {i+1}: {label}")
```
